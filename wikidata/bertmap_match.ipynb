```python
from deeponto.onto import Ontology
from deeponto.align.bertmap import BERTMapPipeline, DEFAULT_CONFIG_FILE
```

```python
def run_deeponto_matching(module:str):
    src_onto_file = f"/kaggle/input/polifonia-music-representation/{module}.owl"
    tgt_onto_file = "/kaggle/input/wikidata-music-ontology/music.owl"
    config_file = "/kaggle/input/polifonia-wikidata-bertmapyaml/bertmap.yaml"
    #config_file = DEFAULT_CONFIG_FILE
    
    config = BERTMapPipeline.load_bertmap_config(config_file)
    
    #BERTMapPipeline.save_bertmap_config(config, "bertmap.yaml")
    
    src_onto = Ontology(src_onto_file)
    tgt_onto = Ontology(tgt_onto_file)
    bertmap = BERTMapPipeline(src_onto, tgt_onto, config)
```

```python
run_deeponto_matching("algorithm-pattern")
```

```python
run_deeponto_matching("bells")
```

```python
run_deeponto_matching("cometa")
```
