```python
from deeponto.onto import Ontology
from deeponto.align.bertmap import BERTMapPipeline, DEFAULT_CONFIG_FILE
import shutil
```

```python
!mkdir -p /kaggle/working/matches
```

```python
def run_deeponto_matching(module:str):
    in_folder = "/kaggle/input"
    out_folder = "/kaggle/working"
    
    src_onto_file = f"{in_folder}/polifonia-music-representation/{module}"
    tgt_onto_file = f"{in_folder}/wikidata-music-ontology/music.owl"
    config_file = f"{in_folder}/polifonia-wikidata-bertmapyaml/bertmap.yaml"
    #config_file = DEFAULT_CONFIG_FILE
    
    config = BERTMapPipeline.load_bertmap_config(config_file)
    
    #BERTMapPipeline.save_bertmap_config(config, "bertmap.yaml")
    
    src_onto = Ontology(src_onto_file)
    tgt_onto = Ontology(tgt_onto_file)
    bertmap = BERTMapPipeline(src_onto, tgt_onto, config)
    
    shutil.move(f"{out_folder}/bertmap/match", f"{out_folder}/matches/{module}")
    
    #shutil.make_archive(f"{out_folder}/{module}", 'zip', f"{out_folder}/bertmap/match")
    #shutil.rmtree(f"{out_folder}/bertmap/match")
```
